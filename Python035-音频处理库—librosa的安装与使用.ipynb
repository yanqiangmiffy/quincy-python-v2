Librosa是一个用于音频、音乐分析、处理的python工具包，一些常见的时频处理、特征提取、绘制声音图形等功能应有尽有，功能十分强大。本文主要介绍librosa的安装与使用方法。

https://blog.csdn.net/zzc15806/article/details/79603994

```python
import librosa
```

## 读取音频

```python
filename='data/00DF400A.wav'
```

```python
# Librosa默认的采样率是22050，如果需要读取原始采样率，需要设定参数sr=None:
y, sr = librosa.load(filename,sr=None)
```

## 绘图显示

```python
import librosa.display
import matplotlib.pyplot as plt
```

```python
plt.figure()
librosa.display.waveplot(y[:4410], sr)
plt.title('Beat wavform')
plt.show()
```

## 绘制频谱图

```python
import librosa
import librosa.display
melspec = librosa.feature.melspectrogram(y, sr, n_fft=1024, hop_length=512, n_mels=128)
logmelspec = librosa.power_to_db(melspec)
plt.figure()
librosa.display.specshow(logmelspec, sr=sr, x_axis='time', y_axis='mel')
plt.title('Beat wavform')
plt.show()
```
